```python
import cv2
import copy
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import keras.utils as image
```

```python
vid = cv2.VideoCapture("driving30seconds.mp4")
model = load_model("imageSegmentation100epochs_filters32_15051354.h5")

while (True):
    # Capture the video frame
    # by frame
    ret, frame = vid.read()

    #resize frame

    frame = cv2.resize(frame, (256,256))
    
    # If needed, convert the frame to grayscale
    # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)'
    if ret:
        input = copy.deepcopy(frame)

        test_image = np.expand_dims(input, axis = 0)      

        prediction = model.predict(test_image)

        prediction = cv2.merge((prediction[0,:,:,0],prediction[0,:,:,1],prediction[0,:,:,2]))

    #display = np.concatenate((input, prediction), axis=1)
        
    # Display the resulting frame
    cv2.imshow('Camera feed', prediction)

    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choice

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
cv2.waitKey(1)
```
